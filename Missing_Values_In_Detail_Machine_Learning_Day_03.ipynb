```python
data["embark_town"].value_counts()
# So we have to replace with the mode
```

# **Now Fill the missing values using Techniques of Sk-Learn**

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
df = sns.load_dataset("titanic")
df
```

```python
df.isnull().sum().sort_values(ascending = False)

# Age has missing values so we will impute values in it using a technique of sk learn
```

```python
# impute age column using simple imputer from sklearn

imputer_technique = SimpleImputer(strategy='median')
df['age'] = imputer_technique.fit_transform(df[['age']])
```

```python
# find missing values sum

df.isnull().sum().sort_values(ascending = False)

# Now the missing values of 'age' has been vanished

```

```python
# univariate imputation : It means to make changes in a column by staying just in that specific column

# multivariate imputation : It means that if we read all the variables and after it , we impute value in 'my desired' column , is called as mutivariant imputation
```

# **Multivariate Imputation**

```python
df = sns.load_dataset('titanic')
df
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
# we have to first enable this kind of imputer

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
# sum of missing values

df.isnull().sum().sort_values(ascending = False)
```

```python
imputer = IterativeImputer(max_iter = 20 , n_nearest_features=5)
df['age'] = imputer.fit_transform(df[["age"]])
```

```python
df.isnull().sum().sort_values(ascending = False)
```

# Imputing values using forward fill ad Backward fill

```python
# load the data set again in order to check the missing values

df = sns.load_dataset("titanic")
df
```

```python
# Using forward fill impute age column
# you just need to add forward fill in the method , in order to fill the values in this manner

df['age'] = df['age'].ffill()
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
# for back ward filling we can jsut write :

df["age"] = df['age'].bfill()

# now the values will be imputed through backward way
```
